### TASK 4

```python
def clean_data(df):
    # Remove leading and trailing whitespace in column: 'Category'
    df['Category'] = df['Category'].str.strip()
    # Change column type to datetime64[ns] for column: 'Date'
    df = df.astype({'Date': 'datetime64[ns]'})
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```

### TASK 5
